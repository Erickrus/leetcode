```python
import numpy as np
import matplotlib.pyplot as plt

# Euler Approximation for Ordinary Differential Equation

# Let's define Newton's cooling law as followings:
# dT/dt = -K*(T-T0)
# Let's assume
K, T0 = 0.5, 20.
def f(T):
    return -K*(T-T0)

def euler_approximation(T1, steps, interval):
    x = [0.]
    y = [T1]
    h = float(interval) / float(steps)
    for i in range(steps):
        T1 = T1 + h * f(T1)
        x.append(x[-1]+h)
        y.append(T1)
    return np.array(x), np.array(y)


# Plot 40->20 and 15->20
x, y = euler_approximation(40., 100, 5)
plt.plot(x, y)
x, y = euler_approximation(15., 100, 5)
plt.plot(x, y)

# Plot dash 20
y1 = np.ones(y.shape) * T0
plt.plot(x, y1, linestyle = '--', c='gray', dash_capstyle='round')

# Add other stuff
plt.xlabel('Time(min)')
plt.ylabel('Temperature(C)')
plt.title('Newton\'s Cooling Law\nwith Euler ODE Approximation')
plt.text(s="Newton's Cooling Law", x=2.7, y=35)
plt.text(s=r'$\frac{dT}{dt} = -K\dot(T-T_0)$', x=2.7, y=33)

plt.text(s="Euler approximation by", x=2.7, y=30)
plt.text(s="$y_{n+1}=y_n+h\dot f(x_n,y_n)$", x=2.7, y=28)

plt.text(s="Surrounding Temperature $T_0=20C$", x=0.1, y=20.8)

plt.text(s="$T_1=40C$", x=0.1, y=39.2)
plt.text(s="$T_2=15C$", x=0.1, y=17.2)

''
```
